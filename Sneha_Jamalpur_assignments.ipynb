# Assignment 2: Load and Analyze the Cars Dataset

### Step 1: Load the dataset
The "cars.csv" dataset is loaded using the pandas library. This step allows us to read the data into a DataFrame, which is easier to analyze.

```python
import pandas as pd
```

```python
# Load the dataset
cars_data = pd.read_csv('cars.csv')
```

### Step 2: Display the number of rows and columns
The `shape` attribute of the DataFrame is used to find out how many rows and columns are present in the dataset.

```python
# Display the number of rows and columns
rows, columns = cars_data.shape
rows, columns
```

### Step 3: Analyze the dataset using summary statistics
The `describe()` function is used to generate statistics for numerical columns, including mean, median, standard deviation, and more.

```python
# Analyze the data using summary statistics
summary_stats = cars_data.describe()
summary_stats
```

### Step 4: Interpret the 25th and 75th percentiles
In this step, we calculate the 25th and 75th percentiles of the dataset. Percentiles help us understand the distribution of data values, with the 25th percentile representing the lower quartile and the 75th percentile representing the upper quartile. We use the quantile function and include the numeric_only=True parameter to exclude non-numeric columns.

```python
# Interpret the 25th and 75th percentiles
percentiles = cars_data.quantile([0.25, 0.75], numeric_only=True)
percentiles
```

# Assignment 3: Data Visualization

### Step 1: Visualize the distribution of mpg
A histogram is plotted to show the frequency distribution of the `mpg` (Miles Per Gallon) column. This helps us understand how `mpg` values are spread across the dataset.

```python
import matplotlib.pyplot as plt
```

```python
# Histogram of mpg
plt.hist(cars_data['mpg'].dropna(), bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Miles Per Gallon (mpg)')
plt.xlabel('Miles Per Gallon (mpg)')
plt.ylabel('Frequency')
plt.show()
```

### Step 2: Analyze the relationship between hp and mpg
A scatter plot is used to visualize the relationship between `hp` (horsepower) and `mpg` (miles per gallon). This shows if there is any trend between these variables.

```python
# Scatter plot of hp vs mpg
plt.scatter(cars_data['hp'], cars_data['mpg'], color='green')
plt.title('Horsepower vs. Miles Per Gallon')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles Per Gallon (mpg)')
plt.show()
```

### Step 3: Visualize the gear column with a bar chart
A bar chart is used to show the frequency of each gear type (`gear` column). This provides a clear view of how many cars have 3, 4, or 5 gears.

```python
# Bar chart of gear counts
gear_counts = cars_data['gear'].value_counts()
gear_counts.plot(kind='bar', color='orange')
plt.title('Gear Counts')
plt.xlabel('Number of Gears')
plt.ylabel('Frequency')
plt.show()
```

### Step 4: Boxplot of wt grouped by cyl
A boxplot is used to visualize the distribution of car weights (`wt`) grouped by the number of cylinders (`cyl`). This helps compare weights across different cylinder types.

```python
import seaborn as sns
```

```python
# Boxplot of wt by cyl
sns.boxplot(x='cyl', y='wt', data=cars_data)
plt.title('Boxplot of Weight by Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Weight (wt)')
plt.show()
```

### Step 5: Correlation heatmap
In this step, we visualize the correlation between numeric variables in the dataset using a heatmap.
1. Select only numeric columns from the dataset using select_dtypes(include='number').
2. Compute the correlation matrix using the corr() method.
3. Use sns.heatmap to create a heatmap that displays the correlation values, with color intensities representing the strength of correlations.


```python
# Correlation heatmap
numeric_data = cars_data.select_dtypes(include='number')  # Select only numeric columns
correlation_matrix = numeric_data.corr()  # Compute correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
